Import Libraries

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
#See the data that's been imported
df.head(10)
```

Clean Up the Data

```python
#Filters so only display 50km or 50 miles USA Races from the year 2020
```

```python
#Display 50 Miles races
df[df['Event distance/length'] == '50mi']
```

```python
#Display 50 Kilometer races
df[df['Event distance/length'] == '50km']
```

```python
#Combine 50km with 50mi with isin
df[(df['Event distance/length'].isin(['50km', '50mi']))]
```

```python
#Get events from the year 2020
df[df['Year of event'] == 2020]
```

```python
#Get only USA events
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#Combine everything
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(')) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
df2.head(10)
```

```python
df2.shape
```

Data Cleaning

```python
#Remove USA from event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Clean up athlete age
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove the 'h' from Athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
df2.head()
```

```python
#Drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete age category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

```python

df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
#Remove null values
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#Check for duplicate values
df2[df2.duplicated() == True]
```

```python
#Reset index
df2.reset_index(drop = True)
```

```python
#Check columns data types
df2.dtypes
```

```python
#Convert 'athlete_age' column's data type to Integer
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
#Convert 'Athlete average speed' column's data type to Float
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head(10)
```

```python
#Rename columns
df2 = df2.rename(columns={'Year of event' : 'year',
                          'Event dates' : 'race_day',
                          'Event name' : 'race_name',
                          'Event distance/length' : 'race_length',
                          'Event number of finishers' : 'race_number_of_finishers',
                          'Athlete performance' : 'athlete_performance',
                          'Athlete gender' : 'athlete_gender',
                          'Athlete average speed' : 'athlete_average_speed',
                          'Athlete ID' : 'athlete_id'})
```

```python
df2.head()
```

```python
#Reorder columns
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_average_speed', 'athlete_performance']]
```

```python
df3.head()
```

Questions From Data

```python
#Difference in speed between male to female
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#What age groups are the best in the 50 miles rac (20+ participant group min)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count > 19').head(15)
```

```python
#What age groups are the worst in the 50 miles rac (10+ participant group min)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count > 9').head(15)
```

```python
#Slower in summer than in winter?
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
df3['race_season'] = df3['race_month'].apply(lambda x: "Winter" if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
